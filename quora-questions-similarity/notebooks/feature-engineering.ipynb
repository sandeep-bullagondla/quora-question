## Feature engineering

```python
features_df = pd.merge(raw_data, preprocessed_questions,
                       left_on="qid1", right_on="qid")\
                .drop(["qid", "question", "clean_text_stem", "text_length_stem"], axis=1)\
                .rename(columns={"clean_text_lemma": "q1_lemma",
                                 "text_length_lemma": "q1_lemma_len"})
features_df = pd.merge(features_df, preprocessed_questions,
                       left_on="qid2", right_on="qid")\
                .drop(["qid", "question", "clean_text_stem", "text_length_stem"], axis=1)\
                .rename(columns={"clean_text_lemma": "q2_lemma",
                                 "text_length_lemma": "q2_lemma_len"})
```

```python
features_df.sample(5)
```

```python
# Word count
features_df["q1_len"] = features_df["question1"].apply(lambda x: len(x.split()))
features_df["q2_len"] = features_df["question2"].apply(lambda x: len(x.split()))
```

```python
# Same type of question?
features_df["same_type_questions"] = features_df.apply(lambda x: x["question1"].split()[0] == x["question2"].split()[0],
                                                       axis=1)
```

```python
# Common words
features_df["num_common_words"] = features_df.apply(lambda x: len(set(x["question1"].split())
                                                                  .intersection(set(x["question2"].split()))),
                                                       axis=1)


features_df["num_common_words_lemma"] = features_df.apply(lambda x: len(set(x["q1_lemma"].split())
                                                                  .intersection(set(x["q2_lemma"].split()))),
                                                       axis=1)
```

### Distance Features

```python
from sklearn.metrics import pairwise_distances
```

```python
with open("../embeddings/bow_embeddings.pkl", "rb") as f:
    bow_embeddings = pickle.load(f)

with open("../embeddings/tfidf_embeddings.pkl", "rb") as f:
    tfidf_embeddings = pickle.load(f)
    
with open("../embeddings/w2v_embeddings.pkl", "rb") as f:
    w2v_embeddings = pickle.load(f)
```

```python
bow_embeddings
```

```python
def pairwise_dist(qid1, qid2, emb, metric):
    
    try:
        d = pairwise_distances(emb[qid1], emb[qid2], metric=metric)[0,0]
    except KeyError:
        d = -1
    
    return d
```

```python
features_df["bow_euclidean_dist"] = features_df.apply(lambda x: pairwise_dist(x["qid1"], x["qid2"],
                                                                              bow_embeddings, "euclidean"),
                                                      axis=1)
features_df["bow_manhattan_dist"] = features_df.apply(lambda x: pairwise_dist(x["qid1"], x["qid2"],
                                                                              bow_embeddings, "manhattan"),
                                                      axis=1)
features_df["bow_cosine_dist"] = features_df.apply(lambda x: pairwise_dist(x["qid1"], x["qid2"],
                                                                           bow_embeddings, "cosine"),
                                                   axis=1)
```

```python
features_df["tfidf_euclidean_dist"] = features_df.apply(lambda x: pairwise_dist(x["qid1"], x["qid2"],
                                                                              tfidf_embeddings, "euclidean"),
                                                      axis=1)
features_df["tfidf_manhattan_dist"] = features_df.apply(lambda x: pairwise_dist(x["qid1"], x["qid2"],
                                                                              tfidf_embeddings, "manhattan"),
                                                      axis=1)
features_df["tfidf_cosine_dist"] = features_df.apply(lambda x: pairwise_dist(x["qid1"], x["qid2"],
                                                                           tfidf_embeddings, "cosine"),
                                                   axis=1)
```

```python
w2v_embeddings = {k:v.reshape(-1,1) for k, v in w2v_embeddings.items()}
```

```python
features_df["w2v_euclidean_dist"] = features_df.apply(lambda x: pairwise_dist(x["qid1"], x["qid2"],
                                                                              w2v_embeddings, "euclidean"),
                                                      axis=1)
features_df["w2v_manhattan_dist"] = features_df.apply(lambda x: pairwise_dist(x["qid1"], x["qid2"],
                                                                              w2v_embeddings, "manhattan"),
                                                      axis=1)
features_df["w2v_cosine_dist"] = features_df.apply(lambda x: pairwise_dist(x["qid1"], x["qid2"],
                                                                           w2v_embeddings, "cosine"),
                                                   axis=1)
```
